```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('tested.csv')
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', 1000) 
print(titanic.head())
```

```python
# avg value or mean value
mn = np.mean(titanic['Age'])   
```

```python
#  This is for Mean/Avg value
max_age = int(titanic['Age'].max())
sns.histplot(x= 'Age', data = titanic, bins= [i for i in range(0,max_age+10, 10)])
plt.plot([mn for i in range(0, 130)], [i for i in range(0,130)], c ='red')
# plt.plot([md for i in range(0, 130)], [i for i in range(0,130)], c ='blue')
plt.show()
```

```python
# Getting/checking null value
titanic.isnull().sum()
```

```python
# Fill null values using mean
titanic['Fare'].fillna(titanic['Fare'].mean())
titanic.isnull().sum()
```

```python
mn_fare = np.mean(titanic['Fare'])
md_fare = np.median(titanic['Fare'].fillna(titanic['Fare'].mean()))
print(md_fare)
print(mn_fare)
```

```python
sns.histplot(x= 'Fare', data = titanic, bins= [i for i in range(0,81, 10)])
plt.plot([mn_fare for i in range(0, 130)], [i for i in range(0,130)], c ='red',label= 'mean')
plt.plot([md_fare for i in range(0, 130)], [i for i in range(0,130)], c ='green',label= 'median')
plt.legend()
plt.show()
```
